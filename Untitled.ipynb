```python
df=df.sample(10000).drop('link',axis=1).reset_index(drop=True)
```

```python
df['text']=df['text'].str.lower().replace(r'^\w\s', '',regex=True).replace('r\n','',regex=True)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.data.path.append('C:\\nltk_data')  # Replace with the path where your punkt data is located

from nltk.stem.porter import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
def token(txt):
    token=nltk.word_tokenize(txt)
    a=[stemmer.stem(w)for w in token]
    return " ".join(a)
```

```python
token ("you are beautiful , beauty")
```

```python
df['text'].apply(lambda x: token(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf= TfidfVectorizer(analyzer='word',stop_words='english')
```

```python
matrix=tfidf.fit_transform(df['text'])
```

```python
matrix
```

```python
similarity=cosine_similarity(matrix)
```

```python
similarity[0] 
```

```python
df[df['song']=="Waiting For The Man"] #Gives the song ID
```

```python
df.tail(5)
```

```python
def recommender(song_name):
    idx= df[df['song']==song_name].index[0]
    distance= sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x:x[1])
    song= []
    for s_id in distance[1:5]:
        song.append(df.iloc[s_id[0]].song)
    return song
```

```python
recommender("Waiting For The Man")
```
